```python
import matplotlib.pyplot as plt

# Metrics from your data
video_names = ["VID1", "VID2", "VID3", "VID4", "VID5"]
TP_values = [52, 266, 1244, 152, 314]
FP_values = [0, 21, 0, 0, 69]
FN_values = [59, 12, 32, 15, 20]

# Calculate total instances (TP + FP + FN) for each video
total_instances = [TP + FP + FN for TP, FP, FN in zip(TP_values, FP_values, FN_values)]

# Convert to probabilities
TP_probs = [TP / total for TP, total in zip(TP_values, total_instances)]
FP_probs = [FP / total for FP, total in zip(FP_values, total_instances)]
FN_probs = [FN / total for FN, total in zip(FN_values, total_instances)]

# Create a bar chart with cream colors
plt.figure(figsize=(10, 6))
plt.bar(video_names, TP_probs, label="True Positives", color="#06d6a0")
plt.bar(video_names, FP_probs, bottom=TP_probs, label="False Positives", color="#ef476f")
plt.bar(video_names, FN_probs, bottom=[TP + FP for TP, FP in zip(TP_probs, FP_probs)], label="False Negatives", color="#ffd166")

# Add labels and title
plt.xlabel("Video")
plt.ylabel("Probability")
plt.title("Probabilities for Rider Detection Metrics")
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Metrics from your data (summed across videos)
total_TP = 52 + 266 + 1244 + 152 + 314
total_FP = 0 + 21 + 0 + 0 + 69
total_FN = 59 + 12 + 32 + 15 + 20

# Create labels for the pie chart
labels = [f"True Positives Frames: {total_TP}", f"False Positives Frames: {total_FP}", f"False Negatives Frames: {total_FN}"]

# Combine the metrics
metrics = [total_TP, total_FP, total_FN]

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(metrics, labels=labels, autopct="%1.1f%%", colors=["#06d6a0", "#ef476f", "#ffd166"])
plt.title("Combined Metrics for Rider Detection")
plt.axis("equal")  # Equal aspect ratio ensures a circular chart

plt.legend(title="Metrics", loc="upper right")
# Show the plot
plt.show()
```
